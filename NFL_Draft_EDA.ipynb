```python
ml_cpd = cpd_cleaned.drop(columns = ['playerProfileUrl', 'combine60ydShuttle', 
                                     'combineArm', 'combineBench', 'combineWonderlic'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder


X = ml_cpd.drop(columns = ['passyards', 'passtd', 'passint', 'passattempts'])
y = ml_cpd['passyards']

enc = LabelEncoder()
X_encoded = X.apply(enc.fit_transform)
    
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size = 0.2, random_state = 8910)


rfr = RandomForestRegressor()

rfr.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

predictions = rfr.predict(X_test)

print("Mean squared error:", mean_squared_error(y_test, predictions))
print("R^2:", r2_score(y_test, predictions))
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gbr = GradientBoostingRegressor()

gbr.fit(X_train, y_train)
```

```python
predictions = gbr.predict(X_test)

print("Mean squared error:", mean_squared_error(y_test, predictions))
print("R^2:", r2_score(y_test, predictions))
```

```python
cpd_cleaned = cpd_cleaned[cpd_cleaned['combinePosition']!= 'WR']
```

```python
# Define thresholds for passing yards and interceptions
high_pass_yards = 5000
moderate_pass_yards = 1000

# Function to categorize players
def categorize_player(row):
    round_drafted = row['round']
    pass_yards = row['passyards']
    pass_int = row['passint']

    if pd.isna(pass_yards):
        pass_yards = 0
    if pd.isna(pass_int):
        pass_int = 0

    if (round_drafted == 1 and pass_yards >= high_pass_yards) or pass_yards >= high_pass_yards:
        return 'Star'
    elif (round_drafted == 1 and moderate_pass_yards <= pass_yards < high_pass_yards) or (round_drafted <= 3 and moderate_pass_yards <= pass_yards < high_pass_yards):
        return 'Meh'
    elif pass_yards < moderate_pass_yards:
        return 'Bust'
    else:
        return 'Meh'

# Apply the function to create the new column
cpd_cleaned['Category'] = cpd_cleaned.apply(categorize_player, axis=1)

# Display the updated dataframe with the new column
cpd_cleaned[['nameFirst', 'nameLast', 'round', 'passyards', 'passint', 'Category']].head()
```

```python
ml_cpd = cpd_cleaned.drop(columns = ['playerProfileUrl', 'combine60ydShuttle', 
                                     'combineArm', 'combineBench', 'combineWonderlic'])
```

```python
from sklearn.ensemble import RandomForestClassifier

X = ml_cpd.drop(columns = ['passyards', 'passtd', 'passint', 'passattempts', 'Category'])
y = ml_cpd['Category']

enc = LabelEncoder()
X_encoded = X.apply(enc.fit_transform)
    
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size = 0.2, random_state = 8910)

rfc = RandomForestClassifier()

rfc.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
predictions = rfc.predict(X_test)

print("Accuracy Score:", accuracy_score(y_test, predictions))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gbc = GradientBoostingClassifier()

gbc.fit(X_train, y_train)
```

```python
predictions = gbc.predict(X_test)

print("Accuracy Score:", accuracy_score(y_test, predictions))
```

```python
# Get feature importances
importances = rfr.feature_importances_

# Get feature names
feature_names = X_train.columns

# Create a DataFrame for better visualization
feature_importances = pd.DataFrame({'Feature': feature_names, 'Importance': importances})

# Sort the DataFrame by importance
feature_importances = feature_importances.sort_values(by='Importance', ascending=False)

# Plot the feature importances
plt.figure(figsize=(10, 6))
plt.barh(feature_importances['Feature'], feature_importances['Importance'])
plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.title('Feature Importance for Random Forest Classifier')
plt.gca().invert_yaxis()
plt.show()
```
